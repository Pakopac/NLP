```python
import numpy
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data_dir = '/content/drive/MyDrive/Colab_Notebooks/'
```

```python
df = pd.read_csv(data_dir + 'movies.csv')
df.head()
```

```python
def combine_features(df):
  df["combined_features"] = df["keywords"] + df["cast"] + df["genres"] + df["director"]
  return df
```

```python
df = df.fillna("")
```

```python
combine_features(df)
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df["combined_features"]) 
X.toarray()
```

```python
sim = cosine_similarity(X)
print(sim)
```

```python
id_titanic = df.index[df['original_title'] == 'Titanic'].tolist()[0]
sim_titanic = dict(enumerate(x for x in sim[id_titanic]))
```

```python
sim_titanic = {k: v for k, v in sorted(sim_titanic.items(), key=lambda item: item[1], reverse=True)}
```

```python
sim_titanic.sort()
sim_titanic = sim_titanic[::-1]
sim_titanic
```

```python
i = 0
movies_similars = []
for val in sim_titanic:
    if(i > 5):
      break
    if(i != 0):
      movies_similars.append(df.iloc[val]["original_title"])
    i += 1
```
